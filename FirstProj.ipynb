**Sample:**

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

**Проверка первых букв двух слов на одинаковость:**

```python
def isLettersInitial (word1, word2):
  if type(word1) != type(word2) and type(word1) == str:
    return False
  elif len(word1.strip()) <= 0 or len(word2.strip()) <= 0:
    return False
  elif word1[0].lower() != word2[0].lower():
    return False
  else:
    return True

print(isLettersInitial('Apple', 'Auntie'))
```

**Вывод всех букв в слове:**

```python
def printAllLetters(word):
   if type(word) != str:
     return False;
   for char in word:
     print(char)


printAllLetters('Apple')
```
